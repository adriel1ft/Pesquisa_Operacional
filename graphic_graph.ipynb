```python
import matplotlib.pyplot as plt
import networkx as nx

# Cria um grafo dirigido (diagraph)
G = nx.DiGraph()

# Adiciona os nós ao grafo
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)

# Adiciona as arestas (conexões entre os nós) para formar a árvore binária
G.add_edge(1, 2)  # Nó 1 aponta para o Nó 2
G.add_edge(1, 3)  # Nó 1 aponta para o Nó 3
G.add_edge(2, 4)  # Nó 2 aponta para o Nó 4
G.add_edge(2, 5)  # Nó 2 aponta para o Nó 5

# Posições dos nós na visualização
pos = nx.spring_layout(G)

# Desenha o grafo
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=12, font_weight='bold', arrows=True)

# Mostra o gráfico
plt.show()
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def print_tree(node, level=0, prefix='Root:'):
    if node is not None:
        print(' ' * (level * 4) + prefix + str(node.value))
        print_tree(node.left, level + 1, 'Left:')
        print_tree(node.right, level + 1, 'Right:')

# Cria a árvore binária
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Imprime a árvore
print_tree(root)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def create_graph(node, pos=None, x=0, y=0, horizontal_distance=1, vertical_distance=2, graph=None):
    if pos is None:
        pos = {}
    if graph is None:
        graph = nx.DiGraph()

    pos[node.value] = (x, y)
    if node.left is not None:
        graph.add_edge(node.value, node.left.value)
        new_x = x - horizontal_distance
        new_y = y - vertical_distance
        create_graph(node.left, pos, new_x, new_y, horizontal_distance / 2, vertical_distance, graph)
    if node.right is not None:
        graph.add_edge(node.value, node.right.value)
        new_x = x + horizontal_distance
        new_y = y - vertical_distance
        create_graph(node.right, pos, new_x, new_y, horizontal_distance / 2, vertical_distance, graph)
    
    return pos, graph

# Cria a árvore binária
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Cria o grafo
pos, graph = create_graph(root)

# Desenha o grafo
nx.draw(graph, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=12, font_weight='bold', arrows=True)

# Mostra o gráfico
plt.show()
```
